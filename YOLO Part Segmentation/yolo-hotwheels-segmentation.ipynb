```python
#Create YAML
yaml_content = """
path: /kaggle/input/hotwheels/Dataset_YOLO
train: images/train
val: images/val
test: images/test

names:
  0: body
  1: wheels
  2: backbody
"""

with open("config.yaml", "w") as f:
    f.write(yaml_content)
```

```python
#Install ultralytics
!pip install ultralytics
```

```python
#Import YOLO
from ultralytics import YOLO
#Choose the model (YOLO11m)
model = YOLO("yolo11m-seg.pt")
```

```python
#Train the model
model.train(
    data="config.yaml",
    epochs=150,
    imgsz=800,
    batch=16,
    device=0,
    project="segmentation_project_2",
    name="yolov11_segmentation",
    verbose=True,
    plots = True,
    fliplr = 0.5,
    flipud = 0.5
)
```

```python
model.val(split='test')
```
